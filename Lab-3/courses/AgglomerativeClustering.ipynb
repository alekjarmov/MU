```python
import numpy as np

X = np.array([[5,3],
    [10,15],
    [15,12],
    [24,10],
    [30,30],
    [85,70],
    [71,80],
    [60,78],
    [70,55],
    [80,91],])
```

```python
import matplotlib.pyplot as plt

labels = range(1, 11)
plt.figure(figsize=(10, 7))
#plt.subplots_adjust(bottom=0.1)
#plt.scatter(X[:,0],X[:,1], label='True Position')
plt.scatter(X[:,0],X[:,1])

for label, x, y in zip(labels, X[:, 0], X[:, 1]):
    plt.annotate(
        label,
        xy=(x, y), xytext=(-3, 3),
        textcoords='offset points', ha='right', va='bottom')
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt

linked = linkage(X, 'single')

labelList = range(1, 11)

plt.figure(figsize=(10, 7))
dendrogram(linked,
            orientation='top',
            labels=labelList,
            distance_sort='descending',
          # distance_sort='ascending',
            show_leaf_counts=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import numpy as np
```

```python
X = np.array([[5,3],
    [10,15],
    [15,12],
    [24,10],
    [30,30],
    [85,70],
    [71,80],
    [60,78],
    [70,55],
    [80,91],])
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
cluster.fit_predict(X)
```

```python
print(cluster.labels_)
```

```python
plt.scatter(X[:,0],X[:,1], c=cluster.labels_, cmap='rainbow')
```

```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import numpy as np
```

```python
customer_data = pd.read_csv('shopping-data.csv')
```

```python
customer_data.shape
```

```python
customer_data.head()
```

```python
data = customer_data.iloc[:, 3:5].values
```

```python
data.shape
```

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10, 7))
plt.title("Customer Dendograms")
dend = shc.dendrogram(shc.linkage(data, method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
cluster.fit_predict(data)
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(data[:,0], data[:,1], c=cluster.labels_, cmap='rainbow')
```

```python
#single linkage
```

```python
cluster2 = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='single')
cluster2.fit_predict(data)

plt.figure(figsize=(10, 7))
plt.scatter(data[:,0], data[:,1], c=cluster2.labels_, cmap='rainbow')
```

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10, 7))
plt.title("Customer Dendograms")
dend = shc.dendrogram(shc.linkage(data, method='single'))
```

```python
# Complete linkage
```

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10, 7))
plt.title("Customer Dendograms")
dend = shc.dendrogram(shc.linkage(data, method='complete'))
```

```python
cluster3 = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='complete')
cluster3.fit_predict(data)

plt.figure(figsize=(10, 7))
plt.scatter(data[:,0], data[:,1], c=cluster3.labels_, cmap='rainbow')
```

```python
# Average linkage
```

```python
cluster4 = AgglomerativeClustering(n_clusters=7, affinity='euclidean', linkage='average')
cluster4.fit_predict(data)

plt.figure(figsize=(10, 7))
plt.scatter(data[:,0], data[:,1], c=cluster4.labels_, cmap='rainbow')
```

```python
plt.figure(figsize=(10, 7))
plt.title("Customer Dendograms")
dend = shc.dendrogram(shc.linkage(data, method='average'))
```

```python
#Choose the best number of clusters
```

```python
from sklearn.metrics import calinski_harabasz_score

results = {}
for i in range(2,11):
    clusterx = AgglomerativeClustering(n_clusters=i, affinity='euclidean', linkage='ward')
    labels=clusterx.fit_predict(data)
    db_index = calinski_harabasz_score(data, labels)
    results.update({i: db_index})
```

```python
plt.plot(list(results.keys()), list(results.values()))
plt.xlabel("Number of clusters")
plt.ylabel("Calinski-Harabasz Index")
plt.show()
```

```python
from sklearn import metrics


results = {}
for i in range(2,11):
    clusterx = AgglomerativeClustering(n_clusters=i, affinity='euclidean', linkage='ward')
    labels=clusterx.fit_predict(data)
    db_index = metrics.silhouette_score(data, labels, metric='euclidean')
    results.update({i: db_index})


```

```python
plt.plot(list(results.keys()), list(results.values()))
plt.xlabel("Number of clusters")
plt.ylabel("Silhouette Score")
plt.show()
```

```python
from sklearn.metrics import davies_bouldin_score


results = {}
for i in range(2,11):
    clusterx = AgglomerativeClustering(n_clusters=i, affinity='euclidean', linkage='ward')
    labels=clusterx.fit_predict(data)
    db_index = davies_bouldin_score(data, labels)
    results.update({i: db_index})
```

```python
plt.plot(list(results.keys()), list(results.values()))
plt.xlabel("Number of clusters")
plt.ylabel("Davies-Bouldin Index")
plt.show()
```
